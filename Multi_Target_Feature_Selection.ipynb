```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
import warnings
import traceback
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("/Users/lenin.s/Desktop/SOLVERMINDS/PROJECTS/SEDGE/DATASETS/all_data_1000.csv")
```

```python
def multi_target_regression(data,target_1=None,target_2=None,target_3=None,target_4=None,target_5=None): 
    
    try:
        target_names =[target_1,target_2,target_3,target_4,target_5]
        targets = [x for x in target_names if x is not None]

        cat_data = data.select_dtypes(include=["object"])
        num_data = data.select_dtypes(include=["int64", "float64"])
        num_data = num_data.fillna(num_data.mean())
        cat_data = cat_data.fillna(cat_data.mode().iloc[0])
        label_en_output = cat_data.apply(LabelEncoder().fit_transform)
        model_data = pd.concat([num_data.reset_index(drop=True), label_en_output.reset_index(drop=True)], axis=1)

        x = model_data.drop(targets, axis=1)
        y = model_data[targets]

        model = MultiOutputRegressor(RandomForestRegressor())
        model.fit(x,y)
        
        feature_importance_equation = pd.Series([])
        feature_df = pd.DataFrame(x.columns)
        feature_df.columns = ['Feature_Name']
        
        df_feature_importance = pd.DataFrame()
        
        for i in range(0,len(targets)):
            feature_importance_equation[i] = pd.Series(model.estimators_[i].feature_importances_, index = x.columns).sort_values()
            df = pd.DataFrame({'Feature_Name':feature_importance_equation[i].index, 'Feature_value_'+str(targets[i]):feature_importance_equation[i].values})
            feature_df = pd.merge(left=feature_df, right=df, how='left', on='Feature_Name')
        return feature_df
    except:
        traceback.print_exc()
```

```python
multi_target_regression(data,'pms','pss','age')
```
